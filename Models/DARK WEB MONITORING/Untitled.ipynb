```python
from xposedornot import XposedOrNot

xon = XposedOrNot()
email = 'ronakagasti38@gmail.com'

# This gives you the "Total Intrusions" and "Risk Level" for your dashboard
analytics = xon.breach_analytics(email)

if analytics.found:
    print(f"--- SECURITY ALERT FOR {email} ---")
    print(f"Total Breaches: {analytics.exposure_count}")
    print(f"Risk Score: {analytics.risk_score}/100")
    print(f"Most Recent Leak: {analytics.first_breach}")
    print(f"Top Exposed Data: {', '.join(analytics.breach_metrics['top_data_types'])}")
else:
    print("✓ No exposure found on the dark web.")
```
